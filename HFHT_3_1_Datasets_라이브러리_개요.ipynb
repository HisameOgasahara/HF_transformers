### dataset 모듈로 HF 리포토리지가 아닌 내가 가진 데이터셋을 열고 훈련/검증용 분리

```python
from datasets import load_dataset

ratio = 80
dataset = load_dataset( #  이 함수는 Hugging Face Datasets 라이브러리를 사용하여 CSV 파일을 로드하고, 데이터를 훈련(train) 및 테스트(test) 세트로 분할합니다.
    path="csv",
    data_files="/content/Writing/src/sample.csv",
    split={
        "train": f"train[:{ratio}%]",
        "test": f"train[{ratio}%:]",
    },
)
dataset
```

```python
# 위 csv파일의 각 줄은 텍스트(부정 또는 긍정적 감정을 나타내는 문장)와 레이블(1은 긍정, 0은 부정)로 구성됩니다. eg."I love this product, it's amazing!",1
```
